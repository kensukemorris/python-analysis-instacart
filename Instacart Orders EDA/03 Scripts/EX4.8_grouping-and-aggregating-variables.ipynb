## Step 4

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
ords_prods_merge.head(10)
```

```python
ords_prods_merge.shape
```

```python
# create the categorical variable based on max orders

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number','max_order']].head(20)
```

## Step 5

```python
ords_prods_merge.groupby(['loyalty_flag']).agg({'prices': ['mean','median','max','min']})
```

The spread are the customer groups seem similar. Regular and new customers seem to also purchase expexsive products. However, the data needs to be rechecked because the a carton of milk from the diary department cannot be USD 99999. 

## Step 6

```python
#mean price per user

ords_prods_merge['avg_spend_per_user'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
ords_prods_merge.head()
```

```python
# make flag based on the aggregated mean per user using loc

ords_prods_merge.loc[ords_prods_merge['avg_spend_per_user'] < 10,'high_spenders'] = "Low spender"
ords_prods_merge.loc[ords_prods_merge['avg_spend_per_user'] >= 10,'high_spenders'] = "High spender"

ords_prods_merge['high_spenders'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id','avg_spend_per_user','high_spenders']].head(20)
```

## Step 7

```python
#Calculate median days since prior order for each user 

ords_prods_merge['median_days_since_prior_ords'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)

ords_prods_merge[['user_id','days_since_prior_order','median_days_since_prior_ords']].head(20)
```

```python
#create flag based on median days since prior order

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_ords'] > 20,'frequent_user'] = "Non-frequent customer"
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_ords'] > 10) & (ords_prods_merge['median_days_since_prior_ords'] <= 20),'frequent_user'] = "Regular customer"
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_ords'] <= 10,'frequent_user'] = "Frequent customer"
                     
ords_prods_merge['frequent_user'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id','days_since_prior_order','median_days_since_prior_ords','frequent_user']].head(20)
```
